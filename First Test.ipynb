#想要把第一個csv檔導入

```python
import pandas as pd

df = pd.read_csv('FirstTest.csv')

df.head()
```

```python
temp = df[['Complex2Name', 'FValue']]
temp
```

#2015的預算資料

```python
type1 = temp.groupby(['Complex2Name'])['FValue'].sum().reset_index()
```

```python
type1['year'] = 2015
type1
```
